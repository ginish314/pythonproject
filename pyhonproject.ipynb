#QUESTION-3 Compare and contrast mutable and immutable objects in Python with examples

Answer:-1. Mutable Objects

Their value can be changed after creation.

we do not need to create a new object when modifying them.

Examples: List, Dictionary, Set, Bytearray

Example: Mutable (List)

numbers = [1, 2, 3]

numbers[0] = 10        # Changing first element

print(numbers)         # Output: [10, 2, 3]

2. Immutable Objects

Their value cannot be changed after creation.

If you try to modify them, a new object is created.

Examples: String, Tuple, Integer, Float, Boolean

Example: Immutable (String)

name = "Hello"
name = name + " World"   
print(name)              # Output: "Hello World"


Here, a new string object is created; the old one remains unchanged.

#QUESTION-5 Explain the concept of type casting in Python with examples

Answer-Type casting is mainly of two types:

#Implicit Type Casting (Automatic)---
In this type, Python automatically converts one data type to another without programmer's involvement.
It usually happens when a smaller data type is converted into a bigger data type to avoid data loss.

Example:

a = 5        # Integer
b = 2.5      # Float

c = a + b    # Python automatically converts 'a' into float
print(c)     # Output: 7.5
print(type(c))  # Output: <class 'float'>



#Explicit Type Casting (Manual)----
In this type, the programmer manually converts one data type into another using built-in functions like:

int() → Converts to integer

float() → Converts to float

str() → Converts to string

list(), tuple(), set() → Convert to respective collection types


#question-6
How do conditional statements work in Python? Illustrate with examples

Conditional statements in Python are used to make decisions in a program. They allow the program to execute a block of code only if a certain condition is true.

Python uses these main conditional statements:

if

elif (else if)

else


#1. if Statement

Used to run a block of code only when a condition is true.

Example:

age = 18
if age >= 18:
    print("You are eligible to vote.")


#2. if-else Statement

If the condition is true, if block executes; otherwise, else block runs.

Example:

number = 10
if number % 2 == 0:
    print("Even number")
else:
    print("Odd number")


#3 if-elif-else Statement

Used when there are multiple conditions to check.

Example:

marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")

#4. Nested if Statement

An if statement inside another if.

Example:

age = 25
is_citizen = True

if age >= 18:
    if is_citizen:
        print("You can apply for a driving license.")
    else:
        print("Only citizens can apply.")
else:
    print("You are underage.")

#5. Using Logical Operators with Conditions

Logical operators like and, or, not help combine multiple conditions.

Example:

age = 20
has_id = True

if age >= 18 and has_id:
    print("Entry allowed")
else:
    print("Entry denied")   

7. Describe the different types of loops in Python and their use cases with examples.

Loops in Python are used to execute a block of code repeatedly. Python mainly has two types of loops:

#1. for Loop

The for loop is used when we know the number of repetitions or when we want to iterate over a sequence like a list, tuple, string, or range.

Example 1: Loop through a list

fruits = ["apple", "banana", "mango"]
for fruit in fruits:
    print(fruit)


#2. while Loop

The while loop is used when the number of iterations is not known in advance.

It keeps running as long as the condition is True.

Example 1:

i = 1
while i <= 5:
    print(i)
    i += 1


Use cases:

Running code until a condition becomes false

Taking user input repeatedly

Creating menus in programs

# 3. Loop Control Statements

Python provides statements to control loop execution:

Statement	Description
break	Stops the loop immediately
continue	Skips the current iteration and moves to the next
pass	Does nothing, used as a placeholder

Example of break:

for i in range(1, 10):
    if i == 5:
        break
    print(i)


Example of continue:

for i in range(1, 6):
    if i == 3:
        continue
    print(i)

# 4. else with Loops

Python allows an else block with loops.
The else part runs only if the loop ends normally (without break).

Example:

for i in range(3):
    print(i)
else:
    print("Loop finished")
